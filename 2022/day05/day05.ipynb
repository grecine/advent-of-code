```python
import numpy as np
import os
import time

np.set_printoptions(threshold=np.inf)
```

```python
def input(fname):
    fname = os.path.join(fname)

    data = []
    with open(fname) as f:
        for line in f:
            data.append(line.replace("\n", ""))#.strip())
    return data  
```

```python
data = input('input.txt')
# data = input('test-input.txt')
```

```python
# data
```

```python
stack_objs = data[:data.index('')-1]
stack_nums = data[data.index('')-1].strip().split(' ')
stack_nums = list(filter(lambda a: a != '', stack_nums))
stack_locs = [data[data.index('')-1].index(s) for s in stack_nums]

stacks = []
for s in stack_locs:
    stack = [l[s] for l in stack_objs]
    stack = list(filter(lambda a: a != ' ', stack))
    print(s, stack)
    stacks.append(stack)
    
print(stacks)
```

```python
moves = []
for m in data[data.index('')+1:]:
    m = m.split(' ')
    moves.append([m[1],m[3],m[5]])
print(moves)
```

```python
for move in moves:
    num = int(move[0])
    source = int(move[1])-1
    dest = int(move[2])-1
        
    tmp = [stacks[source].pop(0) for i in range(num)]
    tmp.reverse()
    tmp.extend(stacks[dest])
    stacks[dest] = tmp
    
    print(stacks)

''.join([s[0] for s in stacks])
```
