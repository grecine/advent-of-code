```python
# cost function
def fuel_used(x, df, part):
    dx = [abs(n-x) for n in df]
    if part==1:
        fuel_used=dx
    elif part==2:
        dx = [int(round(x)) for x in dx] # Fails w/o int only sometimes
        fuel_used=[sum([i+1 for i in range(f)]) for f in dx]
    return sum(fuel_used)
```
